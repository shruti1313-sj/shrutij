```python
# Q1
# create 2 lists from a given list where
# 1st list will contain all the odd numbers from the original list and the 2nd one will contain all the even numbers l=[1,2,3,4,5,6]
```

```python
l=[1,2,3,4,5,6]
odd=[]
even=[]
for i in l:
    if i%2==0:
        even.append(i)
    else:
        odd.append(i)
print(odd)
print(even)
```

```python
# Q5
# wap that can convert a 2D List to 1D list
```

```python
l=[[1,2,3],[4,5,6]]
l1=[]
for i in l:
    for j in i:
        l1.append(j)
print(l1)
```

```python
# Q6
# WAP to remove duplicate items from a list
# l=[1,2,1,2,3,4,5,3,4]
```

```python
l=[1,2,1,2,3,4,5,3,4]
remove=[]
for i in l:
    if i not in remove:
        remove.append(i)
print(remove)
```

```python
# Q7
# WAP to check if a list is in ascending order or not
```

```python
l=[1,2,3,4,6,5,7]
for i in range(len(l)-1):
    if l[i]>l[i+1]:
        print("not in ascending order")
        break
else:
    print("in ascending order")
```
